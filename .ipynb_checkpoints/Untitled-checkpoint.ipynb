```python
from bs4 import BeautifulSoup
from selenium import webdriver 
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import ElementNotVisibleException as NotVisible
import pandas as pd
```

```python
#create a webdriver object and set options for headless browsing
options = Options()
options.headless = True
driver = webdriver.Chrome('./chromedriver',options=options)
```

```python
#uses webdriver object to execute javascript code and get dynamically loaded webcontent
def get_js_soup(url,driver):
    driver.get(url)
    res_html = driver.execute_script('return document.body.innerHTML')
    soup = BeautifulSoup(res_html,'html.parser') #beautiful soup object to be used for parsing html content
    return soup
```

```python
def scrape_categories(driver):
    categories = pd.DataFrame(columns=['category', 'url'])
    url = 'https://www.allrecipes.com/recipes/'
    soup = get_js_soup(url, driver)
    for category in soup.find_all('li', class_='carouselNav__listItem recipeCarousel__listItem'):
        category_name = category.find('div', class_='carouselNav__linkText').text
        category_url = category.find('a', class_='carouselNav__link recipeCarousel__link')['href']
        categories = categories.append({'category':category_name, 'url':category_url}, ignore_index=True)
    return categories
```

```python
#TO DO: FOLLOW LOAD MORE FOR INFINITE SCROLE ON WEBPAGE TO GATHER ALL RECIPES

def scrape_recipes(url, driver):
    recipes = pd.DataFrame(columns=['recipe_name', 'recipe_url'])
    soup = get_js_soup(url, driver)
    for recipe in soup.find_all('div', class_='component card card__category'):
        recipe_name = recipe.find('h3', class_='card__title').text
        recipe_url = recipe.find('a', class_='card__titleLink manual-link-behavior')['href']
        recipes = recipes.append({'recipe_name':recipe_name, 'recipe_url':recipe_url}, ignore_index=True)
    return recipes
```

```python
#TO DO: exception handling when ingredients section does not exist. This happens with 'https://www.allrecipes.com/gallery/cream-cheese-dips/'

def scrape_ingredients(url, driver):
    ingredients_df = pd.DataFrame(columns=['ingredient', 'quantity', 'unit'])
    soup = get_js_soup(url, driver)
    for ingredients in soup.find_all('li', class_='ingredients-item'):
        ingredient = ingredients.find('input', class_='checkbox-list-input')['data-ingredient']
        quantity = ingredients.find('input', class_='checkbox-list-input')['data-init-quantity']
        unit = ingredients.find('input', class_='checkbox-list-input')['data-unit']
        ingredients_df = ingredients_df.append({'ingredient':ingredient, 'quantity':quantity, 'unit':unit}, ignore_index=True)
    return ingredients_df
```
